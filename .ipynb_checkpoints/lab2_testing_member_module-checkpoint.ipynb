```python
from club.member.membertypes import player, staff
from club import club
```

```python
### ~ Creating Staff Members ~ ####

# Initializing Zidane as Staff
zidane = staff('Zinedine Zidane', 'France', 5500000, 9, 'Skipper')
# Displaying Current Information
zidane.display()
# Updating Job Title
zidane.updateTitle('Manager')
# Displaying Initial Salary
print("\n  Initial Salary: ", zidane.getSalary())
# Updating Salary
zidane.setSalary(7500000)
# Displaying updated Salary
print("  Updated Salary: ", zidane.getSalary(),"\n")
# Updating Year
zidane.newYear()
# Updating Nationality
zidane.updateNationality('French')
# Displaying updated information
zidane.display()
```

```python
### ~ Creating Players ~ ###

# Initializing Sergio as player
ramos = player('Sergio Ramos', 'Spain', 15000000, 14, 'Right Back', 15)
# Displaying Current Information
ramos.display()
# Displaying Initial Salary
print("\n  Initial Salary: ", ramos.getSalary())
# Updating Salary
ramos.setSalary(20000000)
# Displaying updated Salary
print("  Updated Salary: ", ramos.getSalary(),"\n")
# Updating Year
ramos.newYear()
# Updating Nationality
ramos.updateNationality('Spanish')
# Updating Position
ramos.updatePosition('Center Back')
# Updating Jersey Number
ramos.updateJerseyNum(4)
# Displaying updated information
ramos.display()
```

```python
### ~ Creating Club with a List of Players ~ ###

# Creating another player and staff member to pass in as a list when initializing club
benzema = player('Karim Benzema', 'French', 7920000, 10, 'Striker', 9)
vazquez = staff('Roberto Vazquez', 'Spain', 3200000, 1, 'Goalkeeping Coach')
# Storing players and staff in list and converting players to list format
plist = [ramos.asList(), benzema.asList()]
slist = [zidane.asList(), vazquez.asList()]
# Creating club and passing in player and staff list for initialization
madrid = club.club(plist, slist)
# Displaying club members
madrid.displayMembers(salary=True)
print('###################################################')

### ~ Adding Players and Staff to the Club ~ ###

# Adding and displaying new Players
madrid.addPlayer('Player 1', 'Country 1', 123456, 1, 'Goalkeeper', 1)
madrid.displayPlayers()
# Adding and displaying new staff members
madrid.addStaff('Staff 1', 'Country 1', 654321, 4, 'Goalkeeping Coach')
madrid.displayStaff()
```

```python
### ~ Updating and Removing A Player ~ ###

# Displaying Players with their salary before updating and removing
madrid.displayPlayers(salary=True)
# Making Changes to Player
madrid.updatePlayer('Sergio Ramos', nationality="None", salary=100, yearsWClub=100, position="Left Right Out", jersey=0)
# Removing Player
madrid.removePlayer('Karim Benzema')
print('###################################################')
# Displaying player changes and deletion
madrid.displayPlayers(salary=True)
```

```python
### ~ Updating and Removing A Staff Member ~ ###

# Displaying Staff Members with their salary before updating and removing
madrid.displayStaff(salary=True)
# Making changes to staff member
madrid.updateStaff("Zinedine Zidane", nationality="Home", salary=42, yearsWClub=10000, title="Skipper")
# Removing staff member
madrid.removeStaff("Roberto Vazquez")
print('###################################################')
# Displaying staff member changes and deletion
madrid.displayStaff(salary=True)
```
